```python
import pandas as pd
import re
import tkinter as tk
from tkinter import filedialog, messagebox

def extract_phone():
    # เลือกไฟล์ต้นทาง
    input_path = filedialog.askopenfilename(filetypes=[("Excel files", "*.xlsx")])
    if not input_path:
        return

    try:
        # อ่านข้อมูล
        df = pd.read_excel(input_path, sheet_name="Sheet1")

        # ตรวจสอบว่ามีคอลัมน์นี้ไหม
        if 'Remark Comment' not in df.columns:
            messagebox.showerror("Error", "ไม่พบคอลัมน์ชื่อ 'Remark Comment'")
            return

        # ดึงเบอร์โทร 10 หลักจากข้อความ
        df['เบอร์โทร'] = df['Remark Comment'].astype(str).str.extract(r'\b(0[689]\d{8})\b')

        # เลือกไฟล์ปลายทาง
        output_path = filedialog.asksaveasfilename(
            defaultextension=".xlsx",
            filetypes=[("Excel files", "*.xlsx")],
            title="บันทึกไฟล์เป็น"
        )
        if not output_path:
            return

        # บันทึกไฟล์ใหม่
        df.to_excel(output_path, index=False)
        messagebox.showinfo("สำเร็จ", f"บันทึกไฟล์เรียบร้อยแล้ว:\n{output_path}")

    except Exception as e:
        messagebox.showerror("เกิดข้อผิดพลาด", str(e))


# GUI
root = tk.Tk()
root.title("Extract เบอร์โทรจาก Remark Comment")
root.geometry("400x200")

label = tk.Label(root, text="📥 กดปุ่มเพื่อเลือกไฟล์ Excel และดึงเบอร์โทร 10 หลัก", pady=20)
label.pack()

button = tk.Button(root, text="📂 เลือกไฟล์และเริ่มทำงาน", command=extract_phone, height=2, width=30)
button.pack()

root.mainloop()
```
